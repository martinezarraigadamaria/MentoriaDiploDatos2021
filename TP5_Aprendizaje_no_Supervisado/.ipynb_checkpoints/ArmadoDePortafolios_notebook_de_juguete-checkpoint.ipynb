El presente es un notebook de juguete para explorar el armado de portafolios equally weighted con y sin rebalanceo.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sys

if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")
```

```python
def get_naive_portfolio(returns, initial_cap =  100):
    
    performance = returns + 1
    performance.iloc[0] = initial_cap
    performance = performance.cumprod()
    
    return performance
```

#### Cargamos la data

```python
df = pd.DataFrame()

for coin in ['BTC', 'ETH']:

    data = pd.read_csv(f'../data/{coin}USDT.csv').rename(columns={'Unnamed: 0' : 'date'}).set_index('date')
    data['pct_close'] = data[['close']].pct_change()
    df[f'{coin}'] = data.pct_close

df = df.loc['2021-01':]
df.index = pd.to_datetime(df.index)
```

```python
df.head()
```

#### ¿Cómo armar un portafolio naive con N stocks? Sin rebalanceo

Al ser un portafolio naive, repartiremos nuestro capital en N (=2) partes, supongamos que contamos con un capital de 100 USDT, entonces destinaremos la mitad en BTC y ETH.

**Importante** Aquí no tendremos en cuenta las comisiones por comprar/vender un activo

```python
#Armamos el vector de pesos, [w0, w1, w2, ...]
number_of_stocks = len(df.columns)

#vector de pesos, en nuestro ejemplo (1/2, 1/2)
weights = np.ones(number_of_stocks)/number_of_stocks
```

```python
weights
```

```python
# Dados los retornos de cada activo, calculamos los retornos del portfolio.
df_returns = (df * weights).sum(axis = 1)
df_returns = pd.DataFrame(data = df_returns, columns = ['Naive'])
```

```python
# La performance de nuestro portfolio será el siguiente
performance_portfolio_naive = get_naive_portfolio(df_returns)
performance_btc = get_naive_portfolio(df.BTC)
performance_eth = get_naive_portfolio(df.ETH)
```

```python
# Ploteamos nuestro portafolio diversificado vs. el resto de activos.
performance_portfolio_naive.plot(figsize=(20,5))
performance_btc.plot(figsize=(20,5))
performance_eth.plot(figsize=(20,5))
plt.legend()
```

#### ¿Cómo armar un portafolio naive con N stocks? Con rebalanceo mensual 

En el caso de nuestro ejemplo, comenzamos con 100 USDT e invertimos 50 USDT en BTC y 50 USDT en ETH. Al finalizar el mes de enero, contamos con 144.410677 USDT en nuestra cartera. Si vemos la performance de BTC y ETH en el mismo periodo, veremos que BTC creció un 14% y ETH un 78% aproximadamente, es decir, crecieron en forma dispar, como suele suceder. Esto implica que nuestra cartera ya no está conformada por 50% de una moneda y 50% de la otra, sino que los "nuevos pesos" son 0.39 y 0.61, para BTC y ETH respectivamente. 

Para comenzar febrero con los mismos pesos (50% para cada par, vamos a rebalancear nuestra cartera). La forma óptima sería vender el 0.11 sobrante de ETH (Tenemos un 61% de nuestra cartera en ETH y queremos llegar a 50%), esto es para reducir lo máximo posible las comisiones. Como en este análisis no tendremos en cuenta dichas comisiones, podemos pensar en pasar nuestro portafolio a USDT, con el "nuevo capital inicial" invertir 50% en BTC y 50% en ETH.

```python
get_naive_portfolio(df_returns.loc['2021-01']).tail()
```

```python
get_naive_portfolio(df[df.columns].loc['2021-01'], round(100 * weights[0], 2)).tail()
```

```python
def get_rebalance_portfolio(df, initial_cap):
    
    #Hard-code !
    
    performance_ = pd.DataFrame()
    for month in range(1,6):
        
        performance = get_naive_portfolio(df[df.columns][(df.index.year==2021)&(df.index.month==month)], initial_cap)
        initial_cap = performance.iloc[-1].mean()
        performance_ = pd.concat([performance_, performance], axis=0)
    
    return pd.DataFrame(performance_.sum(axis=1))
```

```python
performance_portfolio_reb = get_rebalance_portfolio(df, round(100 * weights[0], 2))
performance_portfolio_reb.columns = ['Reb']
```

```python
all_performances = pd.DataFrame()
all_performances['Reb. Port.'] = performance_portfolio_reb.Reb
all_performances['Naive Port.'] = performance_portfolio_naive.Naive
all_performances['BTC'] = performance_btc
all_performances['ETH'] = performance_eth
```

```python
all_performances.plot(figsize=(20,5))
```

En este caso no es notable la diferencia entre el portfolio rebalanceado y el naive, sin embargo, si agregan otras monedas como BNB, podrán ver diferencias.
